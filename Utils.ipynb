```python
fname="KNN/train.txt"
lines = [line.rstrip('\n') for line in open(fname)]
```

```python
id=1
trlabels = []
for s in lines:
    temp = s.split(",")
    for col in range(1,31):
        row = "train("+str(id)+","+str(col)+","+temp[col-1]+")."
        print(row)
    trlabels.append(temp[len(temp)-1])
    id=id+1
```

```python
id=1
for s in lines:
    v = "labels("+str(id)+","+str(trlabels[id-1])+")."
    print(v)
    id=id+1
```
